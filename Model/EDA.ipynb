# **Визуализация данных**

**Одномерный анализ(Распределение переменных):**

```python
fig, axes = plt.subplots(3, 3, figsize=(20, 15)) #В один график объединим несколько подграфиков
sns.histplot(HeartAttack_df['age'], kde=True, ax=axes[0, 0])
axes[0, 0].set(title = 'Распределение по возрасту',
               xlabel = 'Возраст',
               ylabel = 'Количество людей')

sns.countplot(x='sex', data=HeartAttack_df, ax=axes[0, 1])
axes[0, 1].set(title ='Распределение по полу (0 = Женщины, 1 = Мужчины)',
               xlabel = 'Пол',
               ylabel = 'Количество людей')

sns.histplot(HeartAttack_df['chol'], kde=True, ax=axes[1, 0])
axes[1, 0].set(title = 'Распределение по уровню холестерина',
               xlabel = 'Уровень холестерина',
               ylabel = 'Количество людей')

sns.histplot(HeartAttack_df['trtbps'], kde=True, ax=axes[1, 1])
axes[1, 1].set(title = 'Распределение по кровяному давлению',
               xlabel = 'Значение кровянного давления в покое',
               ylabel = 'Количество людей')

sns.histplot(HeartAttack_df['thalachh'], kde=True, ax=axes[2, 0])
axes[2, 0].set(title = 'Распределение по количеству ударов сердца',
               xlabel = 'Mакс. количество ударов сердца',
               ylabel = 'Количество людей')

sns.countplot(x='caa', data=HeartAttack_df, ax=axes[1, 2])
axes[1, 2].set(title ='Распределение по количеству крупных сосудов, окрашенных при рентгеноскопии',
               xlabel = 'Количество сосудов',
               ylabel = 'Количество людей')

sns.countplot(x='fbs', data=HeartAttack_df, ax=axes[0, 2])
axes[0, 2].set(title ='Распределение по количеству сахара в крови > 120 мг/дл (1 = Да, 0 = Нет)',
               xlabel = 'Количество сахара в крови',
               ylabel = 'Количество людей')

sns.histplot(HeartAttack_df['oldpeak'], kde=True, ax=axes[2, 1])
axes[2, 1].set(title = 'Распределение по депрессии сегмента ST',
               xlabel = 'Значение депрессии сегмента ST',
               ylabel = 'Количество людей')

sns.histplot(HeartAttack_df['slp'], kde=True, ax=axes[2, 2])
axes[2, 2].set(title = 'Распределение по наклону пикового сегмента ST',
               xlabel = 'Значение пикового наклона',
               ylabel = 'Количество людей')
plt.tight_layout()
```

Краткий анализ графиков некоторых данных, являющихся одними из первостепенно важных для нас:

1. Распределение по возрасту: Распределен нормально, с концентрацией в возрастной группе 50-60 лет.
2. Распределение по полу: Количество людей мужского пола (обозначено как 1) выше, чем количество людей женского пола (обозначено как 0)
3. Распределение уровня холестерина: Нормальное распределение с концентрацией около 200-300 мг/дл.
4. Распределение артериального давления в состоянии покоя: Несколько нормально распределено, концентрируясь около 120-140 мм рт.ст.

**Корреляционный анализ**

```python
plt.figure(figsize=(20, 15))
sns.heatmap(HeartAttack_df.corr(), annot=True, fmt=".2f", cmap='viridis')
plt.title("Корелляционная матрица")
```

Данный график корреляционной матрицы выявляет несколько примечательных моментов:

1. Существует умеренная отрицательная корреляция между достигнутой максимальной частотой сердечных сокращений() и возрастом(age).
2. Переменная cp (тип боли в груди) показывает умеренную положительную корреляцию с целевой переменной(output), указывая на то, что тип боли в груди может быть значимым предиктором сердечно-сосудистых заболеваний.
3. exng (стенокардия, вызванная физической нагрузкой) и oldpeak (депрессия сегмента ST, вызванная физической нагрузкой относительно покоя) умеренно отрицательно коррелируют с целевой переменной(output).
4. Другие переменные демонстрируют корреляции от низкой до умеренной.

**Анализ распространенности сердечных заболеваний**

```python
heart_disease_count = HeartAttack_df['output'].value_counts()
heart_disease_percentage = heart_disease_count / HeartAttack_df.shape[0] * 100
```

Построим график распространнености:

```python
plt.figure(figsize=(15, 8))
sns.barplot(x=heart_disease_count.index, y=heart_disease_count.values)
plt.title("Распространненость сердечных заболеваний")
plt.xlabel("Наличие сердечного заболевания (0 = Имеется, 1 = Отсутствует)")
plt.ylabel("Количество людей")
plt.xticks([0, 1], ['Сердечное заболевание отсутствует', 'Сердечное заболевание имеется'])

#Процентные значения
for index, value in enumerate(heart_disease_count):
    plt.text(index, value, f'{heart_disease_percentage[index]:.2f}%', ha='center')
```

```python
heart_disease_count, heart_disease_percentage
```

Анализ распространенности сердечных заболеваний показал нам следующее:

165 человек (54,46% от набора данных) имеют сердечные заболевания (обозначены '1' в столбце выходных данных).
У 138 человек (45,54% набора данных) нет сердечных заболеваний (обозначены '0' в столбце выходных данных).

Это указывает на довольно равномерное распределение между людьми с сердечными заболеваниями и без них в нашем dataset'e
